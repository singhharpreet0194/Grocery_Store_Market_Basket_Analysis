```python
max_lift_rows = df.loc[df.groupby('Concatenate(Items (#1))')['Mean(Lift)'].idxmax()]

# Display the resulting DataFrame that contains the rows with the maximum 'Lift' value for each group
max_lift_rows
```

```python
excel_file_path = 'associaltions_data1.xlsx'
max_lift_rows.to_excel(excel_file_path, index=False)
```

```python
max_lift_rows = max_lift_rows.sort_values(by="Mean(Lift)", ascending=False)
```

```python
max_lift_rows.head(20)
```

```python
max_lift_rows1=max_lift_rows[['Concatenate(Items (#1))','Consequent','Mean(Lift)','Mean(Support)','Mean(Confidence)']].sort_values(by="Mean(Lift)", ascending=False)
```

```python
max_lift_rows1.head(20)
```

```python
new_column_names = {
    'Concatenate(Items (#1))': 'Antecedent',
    'Consequent': 'Consequent',
    'Mean(Lift)': 'Lift',
    'Mean(Support)': 'Support',
    'Mean(Confidence)': 'Confidence'
}

# Rename the columns
max_lift_rows1.rename(columns=new_column_names, inplace=True)

max_lift_rows1.head(20)
```
